```python
from lxml import html
import pandas as pd
import re
import os
```

```python
# Shipping to
def getShipTo(tree):
    obs = ''
    for elem in tree.xpath('//table[@class = "shippingTable"]//tr'):
        if len(elem.xpath('//td/label/text()')) > 0:
            obs += ' _%_ '.join(elem.xpath('//td/label//text()')) + ' *_* '
    return obs.strip()
```

```python
# Terms and conditions
def getTermsConditions(tree):
    elem = tree.xpath('(//div[@class = "preformattedNotes"])[2]/pre/text()')[0]
    return elem
```

```python
# Product description
def getProdDesc(tree):
    elem = tree.xpath('(//div[@class = "preformattedNotes"])[1]/pre/text()')[0]
    return elem
```

```python
# Ships from
def getShipFrom(tree):
    try:
        elem = tree.xpath('//div[@class = "tabularDetails"]//label[contains(.,"Ships from")]/following-sibling::span/text()')[0].strip()
    except:
        elem = 'NA'
    return elem
```

```python
# title
def getTitle(tree):
    elem = tree.xpath('//title/text()')[0].strip()
    return elem
```

```python
# exchange rates
def getFxRates(tree):
    elem = '_*_'.join([i.strip() for i in tree.xpath('//div[@class = "exchangeRateListing"]/div/text()')])
    return elem
```

```python
def getVendorName(tree):
    elem =tree.xpath('//div[@class = "tabularDetails"]//a[contains(@href, "contact")]/text()')[0].strip()
    return elem
```

```python
# rating
def getUserRating(tree):
    try:
        elem = tree.xpath('//div[@class = "tabularDetails"]//span[contains(@class,"userRating")]/text()')[0].strip()
    except:
        elem = 'NA'
    return elem
```

```python
import re
# succesful transatcions
def getStatus(tree):
    elem = tree.xpath('//div[@class = "tabularDetails"]//span[@title = "Successful transactions"]/text()')[0]
    elem = re.sub('[(|)]', '', elem)
    return elem
```

```python
# price
def getPrice1(tree):
    elem = tree.xpath('//div[@class = "tabularDetails"]//label[contains(.,"Price")]/following-sibling::span/text()')[0].strip()
    elem = re.sub('[(|)]', '', elem)
    #elem = elem.split(' ')
    return elem
```

```python
# price
def getPrice2(tree):
    elem = tree.xpath('//div[@class = "tabularDetails"]//label[contains(.,"Price")]/following-sibling::div/text()')[0].strip()
    elem = re.sub('[(|)]', '', elem)
    #elem = elem.split(' ')
    return elem
```

```python
# Escrow
def getEscrow(tree):
    elem = tree.xpath('//div[@class = "tabularDetails"]//label[contains(.,"Escrow")]/following-sibling::span/text()')[0].strip()
    return elem
```

```python
def scrapElements(filePath, day):
    with open(filePath, "r") as f:
        page = f.read()
    tree = html.fromstring(page)
    shipTo = getShipTo(tree)
    terms = getTermsConditions(tree)
    prodDesc = getProdDesc(tree)
    shipFrom = getShipFrom(tree)
    title = getTitle(tree)
    rates = getFxRates(tree)
    vendorName = getVendorName(tree)
    rating = getUserRating(tree)
    vendorStatus = getStatus(tree)
    try:
        price = getPrice1(tree)
    except:
        price = getPrice2(tree)
    escrow = getEscrow(tree)
    date = day
    offerId = filePath.split('/')[-1].split('=')[-1]
    return [date, offerId, vendorName, title, price, prodDesc, terms, shipTo, shipFrom, rates, rating, vendorStatus, escrow]
```

```python
def limitFiles(dirList, pat):
    l = []
    for file in dirList:
        if pat in file:
            l.append(file)
    return l
```

```python
offersList = []
basePath = 'E:/dreammarket'
for day in os.listdir(basePath)[111:150]:
    try:
        extendedPath = basePath + '/' + day + '/market'
        offers = limitFiles(os.listdir(extendedPath), 'Product')
        for file in offers:
            try:
                offersList.append(scrapElements(extendedPath + '/' + file, day))
            except:
                print(file)
                print('for day', day, 'didnt scrap anything')
    except:
        extendedPath = basePath + '/' + day
        offers = limitFiles(os.listdir(basePath + '/' + day), 'Product')
        for file in offers:
            try:
                offersList.append(scrapElements(extendedPath + '/' + file, day))
            except:
                print('for day', day, 'didnt scrap anything')
```

```python
df = pd.DataFrame(offersList, columns = ['date', 'offerId', 'vendorName', 'title', 'price', 'prodDesc', 'terms', 'shipTo', 'shipFrom', 'rates', 'rating', 'vendorStatus', 'escrow'])
```
